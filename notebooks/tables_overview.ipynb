### Rozszerzenie danych o `description` 

```python
movies_with_descriptions = pd.read_csv(Path('..','res','data','movies_metadata.csv'))
```

```python
movies_with_descriptions = movies_with_descriptions[['id','title','overview']]
```

```python
movies_with_descriptions.head()
```

```python
movies_with_descriptions.shape
```

```python
movies_without_descriptions = pd.read_csv(Path('..','res','data2','movies.csv'))
movies_without_descriptions.head()
```

```python
movies_without_descriptions['title'] = movies_without_descriptions['title'].str.replace(r'\s*\(\d+\)', '', regex=True)
movies_without_descriptions.head()
```

```python
movies_with_descriptions = movies_with_descriptions[~movies_with_descriptions['title'].isna()]
```

### Poprawiamy tytuły, żeby nie miały ", The" na końcu

```python
mask = movies_without_descriptions['title'].str.endswith(', The')
movies_without_descriptions.loc[mask,'title'] = "The " + movies_without_descriptions.loc[mask,'title'].str.rstrip(", The")
```

```python
mask = movies_without_descriptions['title'].str.endswith(', A')
movies_without_descriptions.loc[mask,'title'] = "A " + movies_without_descriptions.loc[mask,'title'].str.rstrip(", A")
```

```python
mask = movies_without_descriptions['title'].str.endswith(', An')
movies_without_descriptions.loc[mask,'title'] = "An " + movies_without_descriptions.loc[mask,'title'].str.rstrip(", An")
```

```python
movies_without_descriptions.loc[mask,'title'] = "The " + movies_without_descriptions.loc[mask,'title'].str.rstrip(", The")
```

```python
complete_movies_table = movies_without_descriptions.merge(movies_with_descriptions[['title','overview']], 
                                                          on='title', how = 'left')
```
